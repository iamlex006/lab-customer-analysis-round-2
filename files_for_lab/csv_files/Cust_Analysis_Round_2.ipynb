```python
#Response rate by total claim amount. First create categories for the total claim amount, then use same method as above.
```

```python
def claim_cat(x):
    if x <500:
        return "A 0-499"
    elif x<1000:
        return "B 500-999"
    elif x<1500:
        return "C 1000-1499"
    elif x<2000:
        return "D 1500-1999"
    elif x<2500:
        return "E 2000-2499"
    else:
        return "F 2500-2999"
    
data2['Claim Interval'] = list(map(claim_cat, data2['Total Claim Amount']))
data2['Claim Interval'].unique()  # To check the results again
data2
```

```python
table_claim = data2.groupby(["Claim Interval", "Response"]).size()
display(table_claim)
```

```python
rates_claim = []
for i in range(1, len(table_claim), 3):
    rates_claim.append(table_claim[i]/(table_claim[i]+table_claim[i-1]+table_claim[i-2])*100)
rates_claim
```

```python
def income_cat(x):
    if x <20000:
        return "A 0 - 19999"
    elif x<40000:
        return "B 20000 - 39999"
    elif x<60000:
        return "C 40000 - 59999"
    elif x<80000:
        return "D 60000 - 79999"
    else:
        return "E 80000 - 99999"
    
data2['Income Interval'] = list(map(income_cat, data2['Income']))
data2['Income Interval'].unique()  # To check the results again
data2
```

```python
table_income = data2.groupby(["Income Interval", "Response"]).size()
display(table_income)
```

```python
rates_income = []
for i in range(1, len(table_income), 3):
    rates_income.append(table_income[i]/(table_income[i]+table_income[i-1]+table_income[i-2])*100)
rates_income
```
