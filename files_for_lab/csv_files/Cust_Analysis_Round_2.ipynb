```python
data["State"].value_counts(dropna=False)
```

```python
data["State"] = data["State"].fillna("Unknown") #Too many NaN to use "most common value"so creating Ünknown"state. 

data["State"].value_counts(dropna=False)
```

```python
data["Response"].value_counts(dropna=False)
```

```python
data["Response"] = data["Response"].fillna("Unknown") #5% NaN attributed to most common would skew the results so also creating Ünknown"

data["Response"].value_counts(dropna=False)
```

```python
#For both Months Since Last Claim and Open Complaints, since these are quantifiable I'll replace the NaN with the mean of the data set. 
```

```python
data["Months Since Last Claim"].value_counts(dropna=False)
```

```python
import numpy as np
mean_claim = data['Months Since Last Claim'].mean()
mean_claim
data['Months Since Last Claim'] = data['Months Since Last Claim'].fillna(mean_claim)
```

```python
data["Months Since Last Claim"].value_counts(dropna=False) #just checking
```

```python
data["Number of Open Complaints"].value_counts(dropna=False)
```

```python
mean_complaints = data['Number of Open Complaints'].mean()
mean_complaints
data['Number of Open Complaints'] = data['Number of Open Complaints'].fillna(mean_complaints) 
```

```python
data["Number of Open Complaints"].value_counts(dropna=False)
```

```python
data["Vehicle Class"].value_counts(dropna=False)
```

```python
data["Vehicle Class"] = data["Vehicle Class"].fillna("Four-Door Car") #4 door car has 50% of set, so makes sense to add NaN to that cateogry.
data["Vehicle Class"].value_counts(dropna=False)
```

```python
data["Vehicle Size"].value_counts(dropna=False)
```

```python
data["Vehicle Size"] = data["Vehicle Size"].fillna("Medsize") #4 door car has 50% of set, so makes sense to add NaN to that cateogry.
data["Vehicle Size"].value_counts(dropna=False)
```

```python
data["Vehicle Type"].value_counts(dropna=False)
```

```python
data["Vehicle Type"] = data["Vehicle Type"].fillna("Unknown") #I'd be tempted to drop the column as doesn't seem useful, but keeping it in case find use later.
data["Vehicle Type"].value_counts(dropna=False)
```
